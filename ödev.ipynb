```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import matplotlib as mpl

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import warnings;
warnings.filterwarnings("ignore")
```

```python
m = np.arange(0,11)
m
```

```python
c = 3* 10**8
```

```python
E = m*c**2
E
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(10,8))
plt.plot(E, "r")
plt.plot(E, "r")
plt.xlabel("M in g")
plt.ylabel("M in j")
plt.title("E = mc^2")
```

```python
e=np. log(E)
e
```

```python

plt.figure(figsize=(10,8),)
plt.gca().set_yscale('log')

plt.plot(E, "r")
plt.plot(E, "b")
plt.xlabel("M in g")
plt.ylabel("M in j")
plt.title("E = mc^2")
plt.grid(True)
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize=(10,5))
plt.plot(labels,july16_2007,"r",label="july16_2007")
plt.plot(labels,july16_2020,"b",label="july16_2020")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0)
plt.grid(False)

```

```python
plt.figure(figsize=(10,5))
plt.plot(labels,july16_2007,"r",label="july16_2007")
plt.plot(labels,july16_2020,"b",label="july16_2020")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5))
```

```python
fig, ax = plt.subplots(2,1)
ax[0][0].plot(labels,july16_2007,"r")
ax[1][0].plot(labels,july16_2020,"b")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=0)
plt.tight.layout();
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.plot(labels,july16_2007,"r")
plt.title("july16_2007",fontsize=15)

plt.subplot(2,1,2)
plt.plot(labels,july16_2020,"b")
plt.title("july16_2020",fontsize=15)
plt.tight_layout();
```

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(labels, july16_2007,color="black")

ax.set_ylabel("july16_2007",color="black",fontsize=14)

ax2=ax.twinx()

ax2.plot(july16_2020,color="red")
ax2.set_ylabel("july16_2020",color="red",fontsize=14);
```
